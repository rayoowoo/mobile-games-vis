```python
from flask import Flask
from fastparquet import ParquetFile
import pandas as pd
data = pd.read_parquet('games.parq')
```

```python
games = data[['Name', 'Average User Rating', 'Language Count']]
games = games[games['Average User Rating'] != 0.0]
games
```

```python
length = len(games['Name'].index)
print(length)
```

```python
games_list = []
for i in range(length):
    game = {
        'index': i,
        'Name': games['Name'].iloc[i],
        'Average User Rating': games['Average User Rating'].iloc[i],
        'Language Count': games['Language Count'].iloc[i]
    }
    games_list.append(game)
print(games_list)
```

```python
list_of_games = games.reset_index().to_dict('records')
```
